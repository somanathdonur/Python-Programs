```python
def bills(unit):
    if unit<=100:
        bill=100*2

    elif unit<=200:
        bill=100*4
    elif unit>200:
        bill=100*6
    else:
        bill=(100*2)+(100*4)+(100*6)
    return bill
unit=int(input("Enter Your total Units:"))
total=bills(unit)
print("Total BILL Amount:",total)
        
```
