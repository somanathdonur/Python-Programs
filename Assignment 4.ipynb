```python
emp_name=input("Enter employe name:")
emp_salary=float(input("Enter your salary:"))
reating=int(input("Enter your Performance rating(1-5)"))

bonus=0
if reating==5:
    bonus=emp_salary*0.20
elif reating==4:
    bonus=emp_salary*0.15
elif reating==3:
    bonus=emp_salary*0.10
else:
    bonus=0
total_salary=emp_salary+bonus
print("\n Employe Name:",emp_name)
print("\n Employe Salary:",emp_salary)
print("\n Employe Reating:",reating)
print("\n  Employe Bonus:",bonus)
print("\n Total salary:",total_salary)
```
